```python
from statistics import median

def moving_median_(data):
    
    for value in data:
        index=data.index(value)    
        i=median(data[:index+1])           
        yield  i

   
```

```python
data=[12,34,78,23,45,67,89,90,11,10]



y=moving_median_(data)

for med in y:
    print(med)
 
```
